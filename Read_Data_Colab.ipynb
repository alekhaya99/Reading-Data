```
import pandas as pd
import numpy as np
```

```
url = 'https://raw.githubusercontent.com/wesm/pydata-book/2nd-edition/examples/ex1.csv'
df = pd.read_csv(url)
df
```

```
#The following CSV file does not have a header row unlike the previous one
url_1='https://raw.githubusercontent.com/wesm/pydata-book/2nd-edition/examples/ex2.csv'
df = pd.read_csv(url_1)
df
```

```
#In Python we can specify if we want to allow Pandas to assign the default column names
pd.read_csv(url_1,header=None)
```

```
#In Python we can specify if we want to allow Pandas to assign the default column names
pd.read_csv(url_1,names=[1,2,3,4,5])
```

```
#In Python we can specify if we want to allow Pandas to assign the default column names
pd.read_csv(url_1,names=[2,3,4,5])
```

```
#If we want to specify an index column, we can do it as follows
pd.read_csv(url_1,names=[1,2,3,4,5],index_col=3)
```

```
pd.read_csv(url_1,names=[1,2,3,4,5],index_col=1)
```

```
pd.read_csv(url_1,names=[1,2,3,4,5],index_col=4)
```
